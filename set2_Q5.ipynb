```python
from scipy import stats
import numpy as np
from math import *
```

```python
mean_div1 = 5
mean_div2=7
mean = (mean_div1+mean_div2)*45
std_div1 = 3
std_div2 = 4
std = (sqrt(std_div1**2+std_div2**2)*45)
print('mean=',mean,'std=',std)
```

```python
Range = stats.norm.interval(0.95,mean,std)
Range
```

```python
z_score = np.round(stats.norm.ppf(0.05),3)
z_score
```

```python
fifth_percentile_of_profit = mean+(z_score*std)
fifth_percentile_of_profit
```

```python
p_value_makingloss_div1 =  stats.norm.cdf(0,mean_div1,std_div1)
p_value_makingloss_div1
```

```python
p_value_makingloss_div2 =  stats.norm.cdf(0,mean_div2,std_div2)
p_value_makingloss_div2
```
