```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
data = pd.read_csv("/content/BuyerRatio.csv")
data.head()
```

```python
# Null_Hypothesis: All proportional are equal
# Alternate_Hypothesis: All proportional are not equal
```

```python
df = data.iloc[:,1:]
df
```

```python
chi2,p_value,dof,exp=chi2_contingency(df)
```

```python
chi2
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
  print("Reject the null hypothesis")
else:
  print("Accept the null hypothesis")
```
