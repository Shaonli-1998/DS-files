```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
print('The t-score of 95% confidence interval is:', np.round(stats.t.ppf(0.025,df=24),4))
```

```python
print('The t-score of 96% confidence interval is:', np.round(stats.t.ppf(0.02,df=24),4))
```

```python
print('The t-score of 99% confidence interval is:', np.round(stats.t.ppf(0.005,df=24),4))
```
