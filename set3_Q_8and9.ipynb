```python
from scipy import stats
import numpy as np
from math import *
```

```python
p = 0.5
q = 1 - p
err = 0.04
```

```python
z_score_95 = np.round(stats.norm.ppf(0.025),2)
z_score_95
```

```python
sample_size1 = np.round(((z_score_95**2) * p * q)/(err**2),0)
sample_size1
```

```python
z_score_98 = np.round(stats.norm.ppf(0.01),2)
z_score_98
```

```python
sample_size2 = np.round(((z_score_98**2) * p * q)/(err**2),0)
sample_size2
```
