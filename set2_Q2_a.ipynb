```python
from scipy import stats
import numpy as np
```

```python
mean = 38
std= 6
```

```python
z_44 = (44-mean)/std
z_44
```

```python
p_values_greaterthan44 = 1 - stats.norm.cdf(z_44)
p_values_greaterthan44
```

```python
z_38 = (38-mean)/std
z_38
```

```python
p_values_between38and44 = (stats.norm.cdf(z_44)-stats.norm.cdf(z_38))
p_values_between38and44
```
