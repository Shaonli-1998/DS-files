```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import norm
```

```python
data = pd.read_csv("/content/Cars (1).csv")
data.head()
```

```python
data.shape
```

```python
p_mpg_greaterthan_38 = np.round(1-stats.norm.cdf(38,loc=data['MPG'].mean(),scale=data['MPG'].std()),3)
```

```python
p_mpg_greaterthan_38
```

```python
p_mpg_lessthan_40 = np.round(stats.norm.cdf(40,loc=data['MPG'].mean(),scale=data['MPG'].std()),3)
```

```python
p_mpg_lessthan_40
```

```python
p_mpg_lessthan_20 = np.round(stats.norm.cdf(20,loc=data['MPG'].mean(),scale=data['MPG'].std()),3)
```

```python
p_mpg_lessthan_50 = np.round(stats.norm.cdf(50,loc=data['MPG'].mean(),scale=data['MPG'].std()),3)
```

```python
p_mpg_between_20and50 = p_mpg_lessthan_50 - p_mpg_lessthan_20
```
