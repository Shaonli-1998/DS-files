```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv("/content/Cutlets.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
# Null_Hypothesis: The diameter of cutlets is equal in unit A and unit B
# Alternate_Hypothesis: The diameter of cutlets is not equal in unit A and unit B
```

```python
p_value = stats.ttest_ind(data['Unit A'],data['Unit B'])[1]
p_value
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
  print("Reject the null hypothesis")
else:
  print("Accept the null hypothesis")
```
